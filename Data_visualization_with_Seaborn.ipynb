```python
#Importing library
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loaring & Exploring Data

# Load the built-in dataset 'tips'
tips = sns.load_dataset('tips')

# Display the first few rows
print(tips.head())
```

```python
#Basic Plots with Seaborn
#i) Scatter Plot: visulizing the relationship between two variables
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='sex', size='size')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
#ii) Line Plot:

sns.lineplot(data=tips, x='total_bill', y='tip', hue='sex', style='sex', markers=True)
plt.title('Line Plot of Total Bill vs Tip')
plt.show()
```

```python
#iii) Line Plot with Confidence Intervals
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset

#Load the 'iris' dataset
iris = sns.load_dataset('iris')

#Create a line plot standard deviation error bar
sns.lineplot(data=iris, x='sepal_length', y='sepal_width', errorbar='sd', color='blue')

#Add a title and labels
plt.title('Line Plot with Confidence Intervals')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

#Display the plot
plt.show()


```

```python
#iv) Bar Plot:
sns.barplot(data=tips, x='day', y='total_bill', hue='sex', errorbar='sd')
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

```python
#v) Histogram
# sns.histplot(data=tips, x='total_bill', hue='sex', kde=True)
sns.histplot(data=tips, x='total_bill', bins=15, kde=True, color='skyblue') #e.g kernal density estimate overlay
plt.title('Histogram of Total Bill')
plt.show()
```

```python
#vi) Customizing Visulaization: Aesthetics (style)
sns.set_theme(style='whitegrid')  # Options: darkgrid, whitegrid, dark, white, ticks
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Box Plot with Whitegrid Theme')
plt.show()
```

```python
#vii) Color Palettes: Chosing color palette for plots

sns.set_palette('husl')  # Options: deep, muted, bright, pastel, dark, colorblind
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
plt.title('Violin Plot with Husl Palette')
plt.show()
```

```python
#viii) Custom Plots
custom_palette = sns.color_palette(['#2ecc71', '#3498db', '#e74c3c', '#9b59b6'])
sns.set_palette(custom_palette)
sns.countplot(data=tips, x='day', hue='sex')
plt.title('Custom Palette Example')
plt.show()
```

```python
#ix) Context: Adjust the context of plots for presentation or report

sns.set_context('talk')  # Options: paper, notebook, talk, poster
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.title('Scatter Plot with Talk Context')
plt.show()
```

Advanced Plots

```python
#x) Pair Plots: Visualizes relationship between numerical variables in a dataset

sns.pairplot(tips, hue='sex', diag_kind='kde')
plt.show()
```

```python
#xi) Heatmap: Visualizing Correlations or Matrix-like data

# Select only numeric columns
numeric_tips = tips.select_dtypes(include='number')

# Compute the correlation matrix
corr = numeric_tips.corr()

# Create the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#xii) Facet Grid: Allow creating multiple subplots for subsets of the data

g = sns.FacetGrid(tips, col='sex', hue='time', height=4, aspect=1.5)
g.map(sns.scatterplot, 'total_bill', 'tip')
g.add_legend()
plt.show()
```

```python
#xiii) Regression Plot: Combines scatter plots with regression lines
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', markers=['o', 's'])
plt.title('Regression Plot')
plt.show()
```

```python
#Saving plot as figure
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Box Plot')
plt.savefig('box_plot.png', dpi=300, bbox_inches='tight')
```

```python
#Integrating with Pandas
# Explicitly set observed=False to silence the warning in groupby
avg_tip = tips.groupby('day', observed=False)['tip'].mean().reset_index()

# Assign the x variable to hue and set legend=False
sns.barplot(data=avg_tip, x='day', y='tip', hue='day', palette='viridis', dodge=False, legend=False)

# Title and show the plot
plt.title('Average Tip by Day')
plt.show()
```

```python
#Advanced & Unique Visualization with Seaborn
# i) Swarm Plot: Arrange the points to avoid overlapping, ideal for visualizing distribution alongside categories
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', palette='coolwarm')
plt.title('Swarm Plot: Total Bill Distribution by Day and Sex')
plt.show()
```

```python
#ii) Strip plot with Jitter: same as swarm plot but with optional jitter for better visibility
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, jitter=0.3, palette='viridis')
plt.title('Strip Plot with Jitter')
plt.show()
```

```python
#iii) Joint Plot: Scatter plots with histograms (or density plots) for bivariate Analysis

sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex', color='purple')
plt.suptitle('Hexbin Joint Plot: Total Bill vs Tip', y=1.02)
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde', color='blue')
plt.suptitle('KDE Joint Plot: Total Bill vs Tip', y=1)
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg', color='orange')
plt.suptitle('Regression Joint Plot: Total Bill vs Tip', y=1)
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', color='red')
plt.suptitle('Scatter Joint Plot: Total Bill vs Tip', y=1)
plt.show()
```

```python
#iii) Pair Grid with Customization: Pairplot is more customizable version of pairplot
g = sns.PairGrid(tips, hue='sex', palette='husl')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)  # cmap removed
g.map_diag(sns.histplot, kde=True, color='blue')
g.add_legend()
plt.suptitle('Pair Grid: Total Bill, Tip, and Size', y=1.02)
plt.show()
```

```python
g = sns.PairGrid(tips)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap='Blues', fill=True)  # cmap is applied
g.map_diag(sns.histplot, kde=True, color='blue')
g.add_legend()
plt.suptitle('Pair Grid: Total Bill, Tip, and Size', y=1.02)
plt.show()
```

```python
#iv) Custom Heatmaps with Annotations

pivot_table = tips.pivot_table(index='day', columns='time', values='total_bill', aggfunc='mean', observed=False)
sns.heatmap(pivot_table, annot=True, fmt='.2f', cmap='coolwarm', linecolor='w', linewidths=0.5, cbar_kws={'orientation': 'horizontal'})
plt.title('Heatmap of Average Total Bill by Day and Time')
plt.show()
```

```python
#v) Radial Heatmaps (Using Polar Coordinates): Seaborn doesn't support natively polar plots, it combines with matplotlib with Seaborn aesthetics

import numpy as np

# Generate sample data
data = tips.groupby('day', observed=False)['total_bill'].mean().reset_index()
theta = np.linspace(0, 2 * np.pi, len(data), endpoint=False)

# Create a radial heatmap
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
bars = ax.bar(theta, data['total_bill'], color=sns.color_palette('husl', len(data)), alpha=0.8, width=0.5)

# Add labels
ax.set_xticks(theta)
ax.set_xticklabels(data['day'])

# Rotate the labels for better visibility
for label, angle in zip(ax.get_xticklabels(), theta):
    label.set_rotation(np.degrees(angle) - 90)
    label.set_verticalalignment('center')

plt.title('Radial Heatmap: Average Total Bill by Day', va='top')
plt.tight_layout()
plt.show()
```

```python
#vi) FaceGrid with Custom Annotations

g = sns.FacetGrid(tips, col='time', row='sex', margin_titles=True, height=4, aspect=1.5)
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip', hue='day', style='day', palette='tab10')

# Add annotations
for ax in g.axes.flat:
    ax.text(0.5, 0.9, 'Custom Annotation', transform=ax.transAxes, ha='center')

g.add_legend()
plt.suptitle('Facet Grid with Custom Annotations', y=1.02)
plt.show()
```

```python
#vii) Custom KDE Plot

sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True, alpha=0.6, linewidth=2, palette='Set2')
plt.title('Custom KDE Plot: Total Bill by Sex')
plt.show()
```

```python
#viii) Time Series Visualization

# Generate sample time series data
import numpy as np
dates = pd.date_range('2023-01-01', periods=100)
data = pd.DataFrame({'date': dates, 'value': np.random.randn(100).cumsum()})

sns.lineplot(data=data, x='date', y='value', color='red', linewidth=2.5)
plt.title('Time Series Visualization')
plt.xticks(rotation=45)
plt.show()
```

```python
#vii) Custom Spins and Grids
sns.set_theme(style='whitegrid')
ax = sns.boxplot(data=tips, x='day', y='total_bill', hue='day', palette='pastel', dodge=False, legend=False)
sns.despine(offset=10, trim=True)
ax.set_title('Custom Spine Styling')
plt.show()
```

#Matplotlib

```python
#i) Simple Radar Chart

import matplotlib.pyplot as plt
import numpy as np

# Define data
categories = ['Math', 'Reading', 'Writing', 'Science', 'History']
values = [80, 70, 75, 85, 90]

# Create radar chart
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
values += values[:1]  # Repeat the first value to close the circle
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'polar': True})
ax.fill(angles, values, color='blue', alpha=0.25)
ax.plot(angles, values, color='blue', linewidth=2)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_yticklabels(['20', '40', '60', '80', '100'])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

plt.title('Simple Radar Chart')
plt.show()
```

```python
#ii) Multiple Overlapping Radar Charts

# Define data
categories = ['Speed', 'Reliability', 'Comfort', 'Safety', 'Efficiency']
group_data = {
    'Model A': [70, 80, 75, 85, 90],
    'Model B': [60, 70, 80, 75, 65],
    'Model C': [80, 85, 70, 80, 90],
}

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})
for model, values in group_data.items():
    values += values[:1]  # Close the circle
    ax.fill(angles, values, alpha=0.1, label=model)
    ax.plot(angles, values, linewidth=2)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.title('Overlapping Radar Charts')
plt.show()
```

```python
#iii) 3D Radar Chart
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Define data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 40, size=(4, len(categories)))

# Compute angles for radar chart
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False)

# Close the circle for the radar chart
angles = np.concatenate((angles, [angles[0]]))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot each group
for i, value in enumerate(values):
    data = np.concatenate((value, [value[0]]))  # Close the circle for values
    ax.plot(angles, data, zs=i, label=f'Group {i+1}', linewidth=2)

# Set axis ticks and labels
ax.set_xticks(angles[:-1])  # Exclude the last tick for proper labeling
ax.set_xticklabels(categories)  # Match the labels to the number of ticks
ax.set_yticks([10, 20, 30, 40])
ax.set_zticks(range(len(values)))
ax.set_zticklabels([f'Group {i+1}' for i in range(len(values))])

# Add legend and title
ax.legend(loc='upper left', bbox_to_anchor=(1.1, 1.1))
ax.set_title('3D Radar Chart')
plt.show()
```

```python
#Grouped Radar Chart with Transparent Areas
# Define data
categories = ['Skill A', 'Skill B', 'Skill C', 'Skill D', 'Skill E']
values_1 = [75, 80, 65, 70, 85]
values_2 = [65, 70, 75, 80, 60]

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'polar': True})

values_1 += values_1[:1]
values_2 += values_2[:1]

ax.fill(angles, values_1, color='green', alpha=0.25, label='Group 1')
ax.fill(angles, values_2, color='red', alpha=0.25, label='Group 2')

ax.plot(angles, values_1, color='green', linewidth=2)
ax.plot(angles, values_2, color='red', linewidth=2)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.legend(loc='upper right')
plt.title('Grouped Radar Chart')
plt.show()
```
